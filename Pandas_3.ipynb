### use a "groupby" in pandas

```python
import pandas as pd
drinks=pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
drinks.beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks[drinks.continent=='Africa']
```

```python
drinks[drinks.continent=='Africa'].beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.max()
```

```python
drinks.groupby('continent').beer_servings.agg(['count', 'min', 'max', 'mean'])
```

```python
drinks.groupby('continent').mean()
```

### use pandas with scikit-learn to create Kaggle submissions

```python
import pandas as pd
train =pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
feature_cols=['Pclass', 'Parch']
X = train.loc[:, feature_cols]
```

```python
X.shape
```

```python
y = train.Survived
```

```python
y.shape
```

```python
conda install scikit-learn
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X, y)
```

```python
conda list scikit-learn
```

```python
test =pd.read_csv('http://bit.ly/kaggletest')
```

```python
test.head()
```

```python
X_new= test.loc[:, feature_cols]
```

```python
X_new.shape
```

```python
new_pred_class = logreg.predict(X_new)
```

```python
test.PassengerId
```

```python
new_pred_class
```

```python
pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': new_pred_class}).set_index('PassengerId')# dictionary unordered thst's why set index
```
